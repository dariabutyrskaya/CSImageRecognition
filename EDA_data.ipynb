```python
import os
import re
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Define the directory path and regex pattern
path = "Dataset/image_data_fin"
pattern = re.compile(r"(bed|fall|pickup|run|sitdown|standup|walk)")

# Get all .png files in the directory
all_files = [f for f in os.listdir(path) if f.endswith('.png')]

# Dictionary to keep track of the first image for each class
class_images = {}

for f in all_files:
    match = pattern.search(f)
    if match and match.group(1) not in class_images:
        class_images[match.group(1)] = os.path.join(path, f)

# Calculate the number of rows needed
num_classes = len(class_images)
num_rows = (num_classes + 1) // 2  # 2 images per line

# Display an image for each class
fig, axes = plt.subplots(num_rows, 2, figsize=(15, 5 * num_rows))

# Handle case where there's only one row (i.e., axes is a 1D array)
if num_classes <= 2:
    axes = [axes]

for ax, (class_name, image_path) in zip(axes.ravel(), class_images.items()):
    img = mpimg.imread(image_path)
    ax.imshow(img)
    ax.set_title(class_name)
    ax.axis('off')

# If there's an odd number of classes, hide the last axis
if num_classes % 2:
    axes.ravel()[-1].axis('off')

plt.tight_layout()
plt.show()
```
